```python
from pylab import *
from numpy import *
from matplotlib import pyplot as plt



def mandelbrot(c, compl = 0, maxx = 0):
    if abs(compl) > 2 or maxx == 100:
        return abs(compl)
    else: return mandelbrot(c, compl**2 + c, maxx + 1)

matriz = []
for i in linspace(-2,2,10000):
    lista = []
    for j in linspace(-2,2,10000):
        z = mandelbrot(complex(j,i))
        if z < 2:
            lista.append(z)
        else: lista.append(0)
    matriz.append(lista)
```

```python
plt.figure(figsize=(20,20))
imshow(matriz, cmap='gray')
```
