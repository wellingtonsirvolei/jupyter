```python
from pylab import *
from numpy import *
from matplotlib import pyplot as plt

# Função Mandelbrot.
def mandelbrot(c):
    z = 0
    n = 0
    while abs(z) <= 2 and n <= 100:
        z = z*z + c
        n += 1
    return n # Retorna o numero de interações

matriz = []
for i in linspace(-2,2,1000):
    lista = []
    for j in linspace(-2,2,1000):
        z = mandelbrot(complex(j,i))
        if z >= 100: lista.append(0)
        else: lista.append(1)
    matriz.append(lista)
```

```python
plt.figure(figsize=(20,20))
imshow(matriz, cmap='gray')
show()
```
